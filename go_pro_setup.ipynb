```python
import requests

camera_ip = "http://10.5.5.9:8080"
```

```python
# Regular keep alives is necessary sometimes.

response  = requests.get(camera_ip + "/gopro/camera/keep_alive")
print(response.status_code)
print(response.text)
```

```python
import time
while True:
    time.sleep(3)
    response  = requests.get(camera_ip + "/gopro/camera/keep_alive")
    print(response.status_code)
    print(response.text)
```

```python
response  = requests.get(camera_ip + "/gopro/camera/stream/start")
print(response.status_code)
print(response.text)
```

```python
# Start Camera Stream

response  = requests.get(camera_ip + "/gopro/camera/stream/start")
print(response.status_code)
print(response.text)
```

```python
# Stop Camera Stream

response  = requests.get(camera_ip + "/gopro/camera/stream/stop")
print(response.status_code)
print(response.text)
```
